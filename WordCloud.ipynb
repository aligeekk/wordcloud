```python
from wordcloud import WordCloud
from collections import Counter
from PIL import Image
import numpy as np
import pandas as pd

from nltk import word_tokenize
import re

import matplotlib.pyplot as plt
```

```python
# Adjust size in preview to match size needed 
# Use black and white image
img_mask = np.array(Image.open( "./tmp.jpg"))
```

```python
df =pd.read_csv("./test_nlp.txt", header=None, names=['id','value'])
value_tot = np.sum(df.value)
df
```

```python
df_dict=df.set_index('id')['value'].to_dict()
```

```python
cnt = Counter()
for key in df_dict:
    cnt[key] = df_dict[key]/value_tot
```

```python
wordcloud = WordCloud(background_color="white", max_words=500, mask=img_mask, max_font_size=45, random_state=30).generate_from_frequencies(cnt)
```

```python
#https://www.w3schools.com/colors/colors_hsl.asp
def red_color_func(word, font_size, position,orientation,random_state=None, **kwargs):
    return("hsl(215,100%%, %d%%)" % np.random.randint(30,60))

wordcloud.recolor(color_func = red_color_func)
```

```python
plt.figure(dpi=800)
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```
